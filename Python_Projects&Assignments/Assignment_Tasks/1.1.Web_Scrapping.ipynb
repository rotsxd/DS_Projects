# Web Scrapping using BeautifulSoup 
In this demonstration we are extracting Europe's healthcare data from wikipedia, structuring it, cleaning it, and validating the gathered data. 

```python
#Importing the required libraries
import numpy as np
import pandas as pd
import urllib.request as urllib2
#import requests as r
import re
import codecs
from bs4 import BeautifulSoup
import codecs
```

```python
#we are looking to import table containing the healthcare data
#finding the 'class' attribute value of the element containing the table, element starts with tag 'table' 
url='https://en.wikipedia.org/wiki/Healthcare_in_Europe'
res_obj=urllib2.urlopen(url)
htmlpage_as_list=res_obj.readlines()#gives a list of htmlpage code i.e content lines in byteform
#we read line by line so as to search each line for required element's tag and attribute

res_obj.close()

htmlpage_as_list
```

```python
#searching through the htmlpage for tags
lst=[]
for lines in htmlpage_as_list:
    #lines=codecs.decode(lines)# this does the same job as str(byte_obj,'UTF-8')
    #lines=lines.strip()
    lines=str(lines,'UTF-8').strip()
    if re.search('table class', lines):
        lst.append(lines)

print(type(lst[0]))
lst
```

```python
lst[0]
```

```python
#extracting the table name using MatchObject
table_title=re.search('"([^"]*)"',lst[0])
table_title.group(1)
```

```python
#extracting table name from all possible matches
table_search=re.findall('"([^"]*)"',lst[0])
table_title=table_search[0]
table_title
```

```python
#using BeautifulSoup after learning title of table, to extract data and create a dataframe using pandas
#res=r.get(url)
res_obj=urllib2.urlopen(url)
htmlpage_as_text=res_obj.read() # entire content as a single string document #r.content for byte form, r.text in unicode 
HCE=BeautifulSoup(htmlpage_as_text,"html5lib")
res_obj.close()
```

```python
#reading html page code in BS container
HCE.contents[1]
```

```python
#getting required table element from the webpage
table=HCE.find('table', attrs={'class':table_title})
table
```

```python
#getting table headers
table_headers=[x.text for x in table.find_all('th')] # .text gives attribute gives content/value part
table_headers
```

```python
#getting all table rows and data
test1=table.find_all('tr')
test1
```

```python
#creating 2D array of data values
rows=[]
for row in table.find_all('tr'): #iterate over all rows and then process their data
    rows.append([data.text.encode('utf-8').decode() for data in row.find_all('td')]) #encoding is important to ensure accurate data representation
rows   
    
```

```python
#have got column labels and rows, now we build dataframe
df1=pd.DataFrame(rows,columns=table_headers)
df1
```

```python
#data cleaning function
def preproc(dat):
    dat.columns=dat.columns.str.replace('\n','')
    dat.dropna(axis=0, how='all', inplace=True)
    dat.replace(['\n'],[''],inplace=True,regex=True )
    dat.replace(['\s\*$'],[''],inplace=True,regex=True )
    dat.replace(['^\s'],[''],inplace=True,regex=True )
    dat.replace([','],[''],inplace=True,regex=True )
    dat.replace(r'\b[a-zA-Z]\b', np.nan, inplace=True, regex=True)
    dat=dat.apply(pd.to_numeric, errors='ignore')
    return dat
```

```python
#cleaning the df
df1=preproc(df1)
df1
```

```python
#chekcing for right datatypes
df1.dtypes
```

```python
#checking for null values
df1.isna().sum(axis=0)
```

## Getting another data from different table

```python
#getting table name in html document
url='https://en.wikipedia.org/wiki/List_of_countries_by_total_health_expenditure_per_capita'
res_obj=urllib2.urlopen(url)#returns a response object
htmlpage_as_list=res_obj.readlines()#byteform lines
res_obj.close()
lst=[]
for line in htmlpage_as_list:
    line=str(line).strip()
    if re.search(r'table class',line):
        lst.append(line)
table_title=re.findall(r'"([^"]*)"',lst[0] )
table_title=table_title[0]

#extracting table data
res_obj=urllib2.urlopen(url)
HEE=BeautifulSoup(res_obj.read(),'html5lib')
res_obj.close()
table=HEE.find('table',{'class':table_title})
headers=[x.text for x in table.find_all('th')]
rows=[]
for row in table.find_all('tr'):
    rows.append([val.text.encode('utf-8').decode() for val in row.find_all('td')])#very important step
```

```python
rows
```

```python
df2=pd.DataFrame(rows,columns=headers)
df2
```

```python
df2=preproc(df2)
df2
```

```python
df2.rename(columns={'Location':'Country'},inplace=True)
df2
```

```python
df2.isna().sum(axis=0)
```

```python
df2.dtypes
```
